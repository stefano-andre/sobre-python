# Potencias con exponente entero

```python
import numpy as np
```

```python
def pow8v1(n):
    return n*n*n*n*n*n*n*n
```

```python
def pow8v2(n):
    nc = n*n
    nq = nc*nc
    return nq*nq
```

```python
import time
import random
```

```python
import math
```

```python
l1 = []
l2 = []
l3 = []
l4 = []
l5 = []
veces = range(100)
for i in veces:
    n = 10000000.0
    
    tic = time.time()
    r1 = pow8v1(n)
    toc = time.time()
    l1.append(toc-tic)
    
    tic = time.time()
    r2 = pow8v2(n)
    toc = time.time()
    l2.append(toc-tic)
    
    tic = time.time()
    r3 = math.pow(n,8)
    toc = time.time()
    l3.append(toc-tic)
    
    tic = time.time()
    r4 = n**8
    toc = time.time()
    l4.append(toc-tic)
    
    tic = time.time()
    r5 = pow(n,8)
    toc = time.time()
    l5.append(toc-tic)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(veces,l1,'r-+',label='v1')
plt.plot(veces,l2,'g-+',label='v2')
plt.plot(veces,l3,'b-+',label='**')
plt.plot(veces,l4,'c-+',label='mapow')
plt.plot(veces,l5,'y-+',label='pypow')
plt.xlabel('index')
plt.ylabel('tiempo')
plt.title('tiempo para cada medici√≥n para n = '+str(n))
plt.legend()
```

```python
l1 = []
l2 = []
l3 = []
l4 = []
l5 = []
start = 10000000
step = start
end = 5*start+step
ns = np.arange(start,end,step).astype(np.float64)
veces = range(100)
for n in ns:
    l1i = []
    l2i = []
    l3i = []
    l4i = []
    l5i = []
    for i in veces:
        tic = time.time()
        r1 = pow8v1(n)
        toc = time.time()
        l1i.append(toc-tic)

        tic = time.time()
        r2 = pow8v2(n)
        toc = time.time()
        l2i.append(toc-tic)

        tic = time.time()
        r3 = math.pow(n,8)
        toc = time.time()
        l3i.append(toc-tic)

        tic = time.time()
        r4 = n**8
        toc = time.time()
        l4i.append(toc-tic)
        
        tic = time.time()
        r5 = pow(n,8)
        toc = time.time()
        l5i.append(toc-tic)
    l1.append(statistics.mean(filtro_mediana(l1i,19)))
    l2.append(statistics.mean(filtro_mediana(l2i,19)))
    l3.append(statistics.mean(filtro_mediana(l3i,19)))
    l4.append(statistics.mean(filtro_mediana(l4i,19)))
    l5.append(statistics.mean(filtro_mediana(l5i,19)))
```

```python
plt.plot(ns,l1,'r-+',label='v1')
plt.plot(ns,l2,'g-+',label='v2')
plt.plot(ns,l3,'b-+',label='**')
plt.plot(ns,l4,'c-+',label='mapow')
plt.plot(ns,l5,'y-+',label='pypow')
plt.legend()
plt.xlabel('numero')
plt.ylabel('tiempo promedio')
plt.title('tiempo promedio vs tamagno')
```
